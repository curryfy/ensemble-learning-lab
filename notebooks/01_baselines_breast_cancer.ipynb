# Stage 1: Breast Cancer Dataset Baselines

Stage 1 starts the ensemble learning lab with a small, built-in binary classification dataset from scikit-learn. We load the data, check the train/test split, review class balance, and compare a first set of baseline models.

```python
from pathlib import Path
import sys

import pandas as pd

project_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.append(str(project_root / "src"))

from ensemble_lab.data import load_breast_cancer_data
```

## Load the data

```python
X_train, X_test, y_train, y_test, feature_names = load_breast_cancer_data()

print(f"Train shape: X={X_train.shape}, y={y_train.shape}")
print(f"Test shape:  X={X_test.shape}, y={y_test.shape}")
print(f"Number of features: {len(feature_names)}")
```

## Class balance

```python
train_balance = pd.Series(y_train, name="target").value_counts(normalize=True).sort_index()
test_balance = pd.Series(y_test, name="target").value_counts(normalize=True).sort_index()

balance = pd.DataFrame({
    "train": train_balance,
    "test": test_balance,
})

balance
```

## Baseline model comparison

Now we train a small group of baseline classifiers and compare them with the same metrics table.

```python
from ensemble_lab.evaluate import evaluate_models
from ensemble_lab.models import get_stage1_models

models = get_stage1_models()
metrics_table = evaluate_models(models, X_train, X_test, y_train, y_test)

metrics_table
```
